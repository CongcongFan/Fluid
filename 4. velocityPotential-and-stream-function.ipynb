```python
import numpy as np
from matplotlib import pyplot as plt
from IPython import display
from matplotlib.animation import PillowWriter
metadata = dict(title='Movie', artist='cong')
# writer = PillowWriter(fps=15, metadata=metadata)
writer = PillowWriter(fps=15, metadata=metadata)
```

```python
x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))

u = 2*x
v = 2*y

vel_potential = x**2+y**2

plt.figure(figsize=(16,8))

plt.quiver(x,y,u,v,label='Velocity Vector')
# plt.contour(x,y,np.sqrt(u**2+v**2),label='Velocity Contour')
plt.contour(x,y,vel_potential)
plt.legend()
plt.colorbar()
```

```python
plt.figure(figsize=(24,12))

plt.quiver(x,y,u,v,label='Velocity Vector')




y_streamline = np.linspace(-10,10,30)
x_streamline = np.exp(np.log(y_streamline)-0.4)
plt.plot(x_streamline,y_streamline)

time=np.arange(0,5,0.1)
x,y = np.meshgrid(np.linspace(-10,10,30),np.linspace(-10,10,30))

x1,y1 = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))
fig = plt.figure(figsize=(24,24))
with writer.saving(fig, "potential.gif", 100):
    
    for t in time:
        
        v = 2*x
        u = 4*np.sin(y+t)

        plt.quiver(x,y,u,v,label='Fluid field')
        # plt.contourf(x,y,np.sqrt(u**2+v**2),cmap='coolwarm')


        # Streamline constant at (2,2)
        xp, yp = 4,7
        C = xp**2+4*np.cos(yp+t)



        y_streamline = np.linspace(-10,10,50)

        x_streamline = np.sqrt(-np.cos(y_streamline+t)+C)

        # print(Gamma)
        plt.plot(x_streamline,y_streamline)

        plt.scatter(0,0,c='black')
        # plt.contourf(x,y,np.sqrt(u**2+v**2))

        plt.xlim(-10,10)
        plt.ylim(-10,10)
        writer.grab_frame()

        plt.cla()
```
