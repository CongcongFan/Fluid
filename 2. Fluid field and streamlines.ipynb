```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

from IPython import display
from matplotlib.animation import PillowWriter
metadata = dict(title='Movie', artist='cong')
# writer = PillowWriter(fps=15, metadata=metadata)
writer = PillowWriter(fps=15, metadata=metadata)
```

https://github.com/codinglikemad/pyAnimTutorial

```python
u0=1
v0=3
omega=1
x0 = 0
y0 = 0
time=np.arange(0,20,0.1)
fig_list=[]
x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))

fig = plt.figure()
    
with writer.saving(fig, "fieldwithContour.gif", 100):
    
    for t in time:
        
        u = u0 * np.sin(omega*(t-y/v0))
        v = v0

        plt.quiver(x,y,u,v,label='Fluid field')
        plt.contourf(x,y,np.sqrt(u**2+v**2),cmap='coolwarm')
        
        
        y_streamline = np.linspace(-10,10,30)

        # Streamline equation
        x_streamline = (np.cos(omega*(t-y_streamline/v0))-np.cos(omega*(t-y0/v0)))*u0/omega+x0

        plt.plot(x_streamline,y_streamline,label='Streamline')
        
        plt.legend()
        plt.scatter(0,0,c='black')
        writer.grab_frame()
        plt.cla()
    plt.colorbar()
```

```python
t = np.pi/(2*omega)
u = u0 * np.sin(omega*(t-y/v0))
v = v0

x0 = 0
y0 = 0

y_streamline = np.linspace(-10,10,30)
x_streamline = (np.cos(omega*(t-y_streamline/v0))-np.cos(omega*(t-y0/v0)))*u0/omega+x0

plt.plot(x_streamline,y_streamline)
plt.quiver(x,y,u,v)

plt.xlim(-2,2)
```
