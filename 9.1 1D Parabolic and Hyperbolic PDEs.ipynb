## 1D explicit (Forward Euler) Method

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from IPython.display import clear_output
import time

plt.rcParams.update({'font.size': 14})
from utilities import plot_phi
```

```python
def applyBC(Nx,phi,t):

    ## Right BC
    phiR = - np.exp(-t)
    phi[-1] = phiR

        
    ## left BC
    i = 0   
    phiL = np.exp(-t)
    phi[i] = phiL

    return phi

def applyIC(Nx,phi,dx):

    for i in range(phi.shape[0]):
        x = i*dx

        phi[i] = np.cos(x)

    return phi
```

```python
start = time.time()
Nx = 21
L = np.pi  # length


dx = L / (Nx - 1)  # Grid size

tol = 1e-6

aE = 1 / dx ** 2
aW = 1 / dx ** 2
a0 = -(2 / dx ** 2)

t=0
alpha = 1
dtmax = (0.5 * dx**2 / alpha)
dt = 0.8* dtmax

phi = np.zeros(Nx)
phi = applyIC(Nx, phi, dx)
phi = applyBC(Nx, phi, t)

```

## Use dt = 0.8* dtmax<img src="img/example5.1.0.png">

```python
# for all the interior node
phi_new1 = np.zeros(phi.shape)
n=1

while n*dt<1.6*dtmax:
    for i in range(1,Nx-1):
        
        phi_new1[i] = phi[i] + (alpha * dt / dx**2) * (phi[i+1] - 2*phi[i] + phi[i-1])

    
    t = dt*n
    phi_new1 = applyBC(Nx,phi_new1,t)

    phi = phi_new1

    n+=1
```

```python
# for all the interior node
phi = np.zeros(Nx)
phi = applyIC(Nx, phi, dx)
phi = applyBC(Nx, phi, t)
phi_new2 = np.zeros(phi.shape)
n=1

while n*dt<9.6*dtmax+dtmax:
    for i in range(1,Nx-1):
        
        phi_new2[i] = phi[i] + (alpha * dt / dx**2) * (phi[i+1] - 2*phi[i] + phi[i-1])

    t = dt*n
    phi_new2 = applyBC(Nx,phi_new2,t)
    phi = phi_new2
    n+=1
```

```python
# for all the interior node
phi_new3 = np.zeros(phi.shape)
phi = np.zeros(Nx)
phi = applyIC(Nx, phi, dx)
phi = applyBC(Nx, phi, t)
n=1

while n*dt<17.6*dtmax+dtmax:
    for i in range(1,Nx-1):
        
        phi_new3[i] = phi[i] + (alpha * dt / dx**2) * (phi[i+1] - 2*phi[i] + phi[i-1])

    t = dt*n
    phi_new3 = applyBC(Nx,phi_new3,t)
    phi = phi_new3
    n+=1
```

```python
# for all the interior node
phi_new4 = np.zeros(phi.shape)
phi = np.zeros(Nx)
phi = applyIC(Nx, phi, dx)
phi = applyBC(Nx, phi, t)
n=1

while n*dt<25.6*dtmax+dtmax:
    for i in range(1,Nx-1):
        
        phi_new4[i] = phi[i] + (alpha * dt / dx**2) * (phi[i+1] - 2*phi[i] + phi[i-1])

    t = dt*n
    phi_new4 = applyBC(Nx,phi_new4,t)
    phi = phi_new4
    n+=1
```

```python
# Analytical solution
x = np.linspace(0,np.pi,Nx)

phi_A = np.zeros(Nx)
for i in range(Nx):

    phi_A[i] = np.exp(-t)*np.cos(x[i])
error = phi_A - phi

plt.figure(figsize=(12,6))
plt.plot(x,phi_A, label='Analytical')
plt.plot(x,phi_new1, label='1.6 dtmax')
plt.plot(x,phi_new2, label='9.6 dtmax')
plt.plot(x,phi_new3, label='17.6 dtmax')
plt.plot(x,phi_new4, label='25.6 dtmax')
plt.plot(x,error, label='Error of 1.6 dtmax')
plt.legend()
plt.xlim([0,3])
```
