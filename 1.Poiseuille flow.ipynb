```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def profile(dp, R, L, mu, r):
    num = dp * R**2 * (1- r**2/R**2)
    den = 4 * L * mu

    return num/den 
```

```python
dp = 100 # pa
R = 20e-3 # outer radius
r = np.arange(-R,R,1e-3)
L = 5 # length of pipe
mu = 0.0091 # viscocity of water
```

```python
solution = profile(dp, R, L, mu, r)
length = np.linspace(0,L,solution.shape[0])
diameter = np.linspace(-R,R,solution.shape[0])
```

```python
solution
```

```python
plt.plot(length,solution)
```

```python
plt.plot(solution,diameter)
plt.xlim(0,solution.max())
plt.ylim(-R,R)
plt.title("A Laminar Poiseuille Flow. Instantaneous Water Velocity Profile in a Pipe")
plt.xlabel('Velocity(m/s)')
plt.ylabel('Pipe radius (m)')

plt.tight_layout()
```
